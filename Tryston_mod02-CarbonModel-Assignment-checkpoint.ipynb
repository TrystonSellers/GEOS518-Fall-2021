### 1. Introduction

In the accompanying notebook [mod02-CarbonModel1.ipynb](./mod02-CarbonModel1.ipynb) and in class we developed a relatively simple model of global carbon balance that assumed that tracked carbon in the atmosphere and biosphere. The system of equations we developed to describe this system are shown below,

$$
\begin{align*}
\frac{dM_1}{dt} &= F_{21} - F_{12}\\
&= k_{21}M_2 - k_{12}M_1
\end{align*}
$$

and,

$$
\begin{align*}
\frac{dM_2}{dt} &= F_{12} - F_{21}\\
&= k_{12}M_1 - k_{21}M_2
\end{align*}
$$

where $M_1$ is the mass of carbon in the atmosphere (GtC), $M_2$ is the mass of carbon in the biosphere (GtC), $F_{12}$ is the flux of carbon from the atmosphere to the biosphere due to photosynthesis (GtC/yr), and $F_{21}$ is the flux of carbon from the biosphere to the atmosphere due to decay and respiration (GtC/yr).

In that notebook we develop a numerical model to describe that system, run it for given initial conditions and with no external inputs of carbon (we call this an "unforced" scenario) and compare it to the analytical solution. We observed that as the time step increased, we departed further from the analytical solution.

## Solution Code

### Problem A: Modify and Rerun the Model

```python
import numpy as np
import matplotlib.pyplot as plt

#Initial Conditions

M1i = 1100.0 #initial mass of carbon in atmosphere (GtC)
M2i = 300.0 # initial mass of carbon in bioshpere (GtC)

k12 = 0.0003 #rate of flux atmosphere to biosphere (1/year)
k21 = 0.1 #rate of flux biosphere to atmosphere (1/year)

ti = 0.0 #initial time (years)
tf = 100.0 #ending time (years)
dt = 1/365 #time step (years)

#Set up containers

t = np.arange(ti,tf+dt,dt) #create an array begining at the inital time, ending at the final time, with a time step of dt
Nt = t.size #determine the size of the array to figure out how many time steps are required for the model

print('t has '+str(Nt)+' time steps') #print the number of time steps required for the model

M1_phot = np.zeros((Nt,)) #create a container for the mass of carbon in the atmosphere
M2_phot = np.zeros((Nt,)) #create a container for the mass of carbon in the biosphere

#Solve system of equations for M1 and M2 at t with photosynthesis dependent on both M1&M2

for i in np.arange(Nt): 
    if (i==0): #if at the initial time, the mass of carbon is equal to the inital values

        M1_phot[i] = M1i 
        M2_phot[i] = M2i
        
    else: #if not at the initial time then the mass needs to be calculated as described at the begining
        dM1dt_phot = k21*M2_phot[i-1] - k12*M1_phot[i-1]*M2_phot[i-1] #calculate the rate of change of M1 at the ith time step
        dM2dt_phot = k12*M1_phot[i-1]*M2_phot[i-1] - k21*M2_phot[i-1] #calculate the rate of change of M2 at the ith time step
        
        M1_phot[i] = M1_phot[i-1] + dM1dt_phot*dt #calculate M1 at the ith time step
        M2_phot[i] = M2_phot[i-1] + dM2dt_phot*dt #calculate M2 at the ith time step
        
print('The final mass of carbon in the atmosphere is '+str(M1_phot[-1])+ 'GtC')
print('The final mass of carbon in the biosphere is '+str(M2_phot[-1])+ 'GtC')
```

### Problem B: Examine impacts of human emissions

```python
#load anthropogenic emissions data for 1751-20008

historical_emissions_file = 'AnthropogenicEmissions.1751_2008.csv'
data = np.loadtxt(historical_emissions_file, delimiter=',', skiprows=1)
year = data[:,0] #years are in the first column
cflux = data[:,1] #emissions are in the second column

#plot anthropogenic emissions data

plt.figure(figsize=(8.5,11)) #create figure for the graphs
plt.rcParams.update({'font.size': 14}) #change the font size
plt.title('Anthropogenic Emissions from 1751 to 2008') #create a title
plt.plot(year,cflux,'b-') #plot the cflux versus the year
plt.ylabel('Anthropogenic Emissions (GtC)') #create x-axis
plt.xlabel('Time (years)') #create y-axis
```

```python
#Create new mass balance equations including anthropogenic emissions

#use final M1 & M2 from part A as new initial conditions

M1i_anth = M1_phot[-1]
M2i_anth = M2_phot[-1]

#need new initial conditions for time
ti_anth = 1751
tf_anth = 2008
dt_anth = 1/365 #time step (years)

#Determine steps required

t_anth = np.arange(ti_anth,tf_anth+dt_anth,dt_anth) #create an array begining at the inital time, ending at the final time, with a time step of dt
Nt_anth = t_anth.size + 1 #determine the size of the array to figure out how many time steps are required for the model (+1 needed for diff calc)

print('t has '+str(Nt_anth)+' time steps') #print the number of time steps required for the model
```

```python
cflux_t = np.interp(t_anth,year,cflux) #interpolate the emissions data for the each time step in the model

#Create new containers

M1_anth = np.zeros((Nt_anth,)) #create a container for the mass of carbon in the atmosphere
M2_anth = np.zeros((Nt_anth,)) #create a container for the mass of carbon in the biosphere

#Solve system of equations

for i in np.arange(Nt_anth): 
    if (i==0): #if at the initial time, the mass of carbon is equal to the inital values

        M1_anth[i] = M1i_anth 
        M2_anth[i] = M2i_anth
        
    else: #if not at the initial time then the mass needs to be calculated as described at the begining
        dM1dt_anth = k21*M2_anth[i-1] - k12*M1_anth[i-1]*M2_anth[i-1] + cflux_t[i-1] #calculate the rate of change of M1 at the ith time step
        dM2dt_anth = k12*M1_anth[i-1]*M2_anth[i-1] - k21*M2_anth[i-1] #calculate the rate of change of M2 at the ith time step
        
        M1_anth[i] = M1_anth[i-1] + dM1dt_anth*dt #calculate M1 at the ith time step
        M2_anth[i] = M2_anth[i-1] + dM2dt_anth*dt #calculate M2 at the ith time step

#need to find diff in carbon for plots 3 & 4

dM1_anth = np.diff(M1_anth)
dM2_anth = np.diff(M2_anth)

#need to realign indices with t
M1_anth = M1_anth[0:i]
M2_anth = M2_anth[0:i]

print('The final mass of carbon in the atmosphere is '+str(M1_anth[-1])+ 'GtC')
print('The final mass of carbon in the biosphere is '+str(M2_anth[-1])+ 'GtC')
```

```python
#plots
        
plt.figure(figsize=(16,20))
plt.rcParams.update({'font.size': 14})

plt.subplot(2,2,1) #plot mass of carbon in the atmosphere versus time
plt.plot(t_anth,M1_anth,'b-')
plt.ylabel('Mass of Carbon in the Atmosphere (GtC)')
plt.xlabel('Time (years)')

plt.subplot(2,2,2) #plot mass of carbon in the biosphere versus time
plt.plot(t_anth,M2_anth,'g-')
plt.ylabel('Mass of Carbon in the Biosphere (GtC)')
plt.xlabel('Time (years)')

plt.subplot(2,2,3) #plot change in mass of carbon in the atmosphere versus time
plt.plot(t_anth,dM1_anth,'b--')
plt.ylabel('Change in Mass of Carbon in the Atmosphere (GtC)')
plt.xlabel('Time (years)')

plt.subplot(2,2,4) #plot change in mass of carbon in the biosphere versus time
plt.plot(t_anth,dM2_anth,'g--')
plt.ylabel('Change in Mass of Carbon in the Biosphere (GtC)')
plt.xlabel('Time (years)')
```
